```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url
```

```python
url =  "https://realty.economictimes.indiatimes.com/news"
response = requests.get(url)
soup = BeautifulSoup(response.text,"html.parser")
```

```python
response.text
```

```python
all_links = []
for link in soup.find_all('a',  attrs={'href': re.compile("^https://")}):
    all_links.append(link.get("href"))
```

```python
len(all_links)
```

```python
from langchain.document_loaders import UnstructuredURLLoader
```

```python
pip install unstructured
```

```python
loader = UnstructuredURLLoader(urls=all_links[10:])
data = loader.load()
```

```python
all_data = []
for each in data:
    if len(each.page_content)>0:
        all_data.append(each.page_content)
```

```python
len(all_data)
```

```python
import pandas as pd
money_control_data = pd.DataFrame(columns=["Money_control_data"],data=all_data)
```
